```python
import pandas as pd

# 加载数据集
file_path = 'github_bot_processed_data.csv'  
data = pd.read_csv(file_path)

# 查看前几行
print(data.head())

# 查看数据基本信息
print(data.info())

# 生成描述性统计信息
print(data.describe())
```

```python
# 将日期列转换为日期时间格式
data['created_at'] = pd.to_datetime(data['created_at'])
data['updated_at'] = pd.to_datetime(data['updated_at'])

# 填充缺失值（
data['bio'] = data['bio'].fillna('No Bio')

# 打印转换后的数据
print("\n日期格式化后的数据前5行：")
print(data.head())
```

```python
import numpy as np

# 对数值列进行log变换以减少偏态
for col in ['public_repos', 'public_gists', 'followers', 'following']:
    if col in data.columns:
        data[f'log_{col}'] = np.log1p(data[col])  # log(1 + x) 防止log(0)报错

# 查看新特征
print("\n添加log变换后的列：")
print(data[['public_repos', 'log_public_repos', 'followers', 'log_followers']].head())
```

```python
import matplotlib.pyplot as plt

# 绘制log_followers的分布
data['log_followers'].hist(bins=30)
plt.title('Log Followers的分布')
plt.xlabel('Log Followers')
plt.ylabel('频率')
plt.show()

# 绘制Site Admin分布的条形图
data['site_admin'].value_counts().plot(kind='bar')
plt.title('Site Admin的分布')
plt.xlabel('Site Admin')
plt.ylabel('数量')
plt.show()

# 绘制public_repos与followers的散点图
plt.scatter(data['public_repos'], data['followers'], alpha=0.5)
plt.title('Public Repos与Followers的关系')
plt.xlabel('Public Repos')
plt.ylabel('Followers')
plt.show()

# 相关性矩阵
corr_columns = ['log_public_repos', 'log_public_gists', 'log_followers', 'log_following']
corr_matrix = data[corr_columns].corr()
plt.matshow(corr_matrix, cmap='coolwarm')
plt.colorbar()
plt.title('相关性矩阵', pad=20)
plt.show()
```

```python
import seaborn as sns

# Boxplot: 根据label查看log_followers的分布
sns.boxplot(x='label', y='log_followers', data=data)
plt.title('不同Label的Log Followers分布')
plt.show()

# Pairplot: 查看特征之间的关系
sns.pairplot(data[['log_public_repos', 'log_public_gists', 'log_followers', 'log_following']])
plt.show()

# 绘制log_followers的分布
sns.histplot(data['log_followers'], kde=True)
plt.title('Log Followers的分布')
plt.show()
```
